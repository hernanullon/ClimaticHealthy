```python
from tslearn.preprocessing import TimeSeriesScalerMinMax
from tslearn.svm import TimeSeriesSVC
from tslearn.metrics import cdist_dtw
from tslearn.preprocessing import TimeSeriesScalerMinMax
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
data_path = '..\data\processed\\cluster_2017_2018[DataMining].csv'
df_data = pd.read_csv(data_path)
df_data['ts'] = pd.to_datetime(df_data['ts'])
df_data = df_data.set_index('ts')
df_data.sample(5)
```

```python
X = df_data.iloc[:,:-1].values
y = df_data.iloc[:,-1].values
```

```python
X_train = df_data.iloc[:365,:-1].values
y_train = df_data.iloc[:365,-1].values
X_test = df_data.iloc[365:,:-1].values
y_test = df_data.iloc[365:,-1].values
```

```python
np.random.seed(0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

X_train = TimeSeriesScalerMinMax().fit_transform(X_train)
X_test = TimeSeriesScalerMinMax().fit_transform(X_test)

clf = TimeSeriesSVC(kernel="gak", gamma=.1,verbose=True)
clf.fit(X_train, y_train)
print("Correct classification rate:", clf.score(X_test, y_test))
```
